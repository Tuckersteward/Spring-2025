```python
import pandas as pd
from sqlalchemy import create_engine

csv_file = 'car_price_dataset.csv'   
database = 'mydatabase'   
table = 'mytable'      
mysql_password = 'my-secret-pw'  

df = pd.read_csv(csv_file)

engine = create_engine(f'mysql+mysqlconnector://root:{mysql_password}@localhost:3306/{database}')

df.to_sql(table, engine, if_exists='replace', index=False)

print(f"Data from {csv_file} successfully imported into {database}.{table}")
```
